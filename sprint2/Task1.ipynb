```polyglot-notebook
using System;

// --- ИСПОЛНЯЕМАЯ ЧАСТЬ ---

// Создаем экземпляр класса Dog (сложное наследование: Animal -> Mammal -> Dog)
Dog myDog = new Dog("Рекс", 5, "Черный", "Немецкая овчарка");
Console.WriteLine("Информация о собаке:");
myDog.DisplayInfo();
myDog.MakeSound();

Console.WriteLine(new string('-', 30));

// Создаем экземпляр класса Cat (простое наследование: Animal -> Cat)
Cat myCat = new Cat("Барсик", 3, true);
Console.WriteLine("Информация о кошке:");
myCat.DisplayInfo();
myCat.MakeSound();

Console.WriteLine(new string('-', 30));

// Создаем экземпляр класса Bird (простое наследование: Animal -> Bird)
Bird myBird = new Bird("Кеша", 2, "Зеленый", true);
Console.WriteLine("Информация о птице:");
myBird.DisplayInfo();
myBird.MakeSound();

// --- КЛАССЫ ---

// Шаг 1: Базовый класс Animal
public class Animal
{
    public string Name { get; set; }
    public int Age { get; set; }

    public Animal(string name, int age)
    {
        Name = name;
        Age = age;
    }

    public virtual void DisplayInfo()
    {
        Console.WriteLine($"Имя: {Name}, Возраст: {Age} лет");
    }

    public virtual void MakeSound()
    {
        Console.WriteLine("Животное издает звук.");
    }
}

// Шаг 2: Промежуточный производный класс Mammal
public class Mammal : Animal
{
    public string FurColor { get; set; }

    public Mammal(string name, int age, string furColor) : base(name, age)
    {
        FurColor = furColor;
    }

    public override void DisplayInfo()
    {
        base.DisplayInfo();
        Console.WriteLine($"Цвет шерсти: {FurColor}");
    }

    public override void MakeSound()
    {
        Console.WriteLine("Млекопитающее издает звук.");
    }
}

// Шаг 3: Производный класс Dog
public class Dog : Mammal
{
    public string Breed { get; set; }

    public Dog(string name, int age, string furColor, string breed) 
        : base(name, age, furColor)
    {
        Breed = breed;
    }

    public override void DisplayInfo()
    {
        base.DisplayInfo();
        Console.WriteLine($"Порода: {Breed}");
    }

    public override void MakeSound()
    {
        Console.WriteLine("Гав-гав!");
    }
}

// Класс Cat
public class Cat : Animal
{
    public bool IsIndoor { get; set; }

    public Cat(string name, int age, bool isIndoor) : base(name, age)
    {
        IsIndoor = isIndoor;
    }

    public override void DisplayInfo()
    {
        base.DisplayInfo();
        Console.WriteLine($"Домашняя кошка: {(IsIndoor ? "Да" : "Нет")}");
    }

    public override void MakeSound()
    {
        Console.WriteLine("Мяу-мяу!");
    }
}

// Класс Bird
public class Bird : Animal
{
    public string Color { get; set; }
    public bool CanFly { get; set; }

    public Bird(string name, int age, string color, bool canFly) : base(name, age)
    {
        Color = color;
        CanFly = canFly;
    }

    public override void DisplayInfo()
    {
        base.DisplayInfo();
        Console.WriteLine($"Окрас: {Color}, Умеет летать: {CanFly}");
    }

    public override void MakeSound()
    {
        Console.WriteLine("Чик-чирик!");
    }
}
```
