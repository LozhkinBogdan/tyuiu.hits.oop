```polyglot-notebook
using System;

// --- ИСПОЛНЯЕМАЯ ЧАСТЬ ---

Console.WriteLine("=== Пример множественного наследования через интерфейсы ===\n");

// Создаем экземпляр класса Dog (реализует ICanBark)
Dog myDog = new Dog("Рекс", 5, "Немецкая овчарка", "Гав-гав");
myDog.DisplayInfo();
myDog.MakeSound();
Console.WriteLine();

// Создаем экземпляр класса Bird (реализует IFlyable)
Bird myBird = new Bird("Кеша", 2, "Зеленый", 1000);
myBird.DisplayInfo();
myBird.MakeSound();
myBird.Fly();
Console.WriteLine();

// Создаем экземпляр класса Duck (реализует ОБА интерфейса: IFlyable и ISwimmable)
Duck myDuck = new Duck("Утка", 3, 500, true);
myDuck.DisplayInfo();
myDuck.MakeSound();
myDuck.Fly();   // Метод из интерфейса IFlyable
myDuck.Swim();  // Метод из интерфейса ISwimmable
Console.WriteLine();

// Создаем экземпляр класса Cat (только базовое наследование)
Cat myCat = new Cat("Барсик", 4, true);
myCat.DisplayInfo();
myCat.MakeSound();

// --- КЛАССЫ И ИНТЕРФЕЙСЫ ---

// Шаг 1: Абстрактный базовый класс Animal
public abstract class Animal
{
    public string Name { get; set; }
    public int Age { get; set; }

    protected Animal(string name, int age)
    {
        Name = name;
        Age = age;
    }

    public abstract void MakeSound();

    public virtual void DisplayInfo()
    {
        Console.WriteLine($"Имя: {Name}, Возраст: {Age} лет");
    }
}

// Шаг 2: Интерфейсы

public interface IFlyable
{
    void Fly();
    int MaxAltitude { get; set; }
}

public interface ISwimmable
{
    void Swim();
    bool IsWaterproof { get; set; }
}

public interface ICanBark
{
    void Bark();
    string BarkType { get; set; }
}

// Шаг 3: Производные классы

public class Dog : Animal, ICanBark
{
    public string Breed { get; set; }
    public string BarkType { get; set; }

    public Dog(string name, int age, string breed, string barkType) : base(name, age)
    {
        Breed = breed;
        BarkType = barkType;
    }

    public override void MakeSound()
    {
        Bark();
    }

    public override void DisplayInfo()
    {
        base.DisplayInfo();
        Console.WriteLine($"Порода: {Breed}");
    }

    public void Bark()
    {
        Console.WriteLine($"{Name} лает: {BarkType}!");
    }
}

public class Bird : Animal, IFlyable
{
    public string Color { get; set; }
    public int MaxAltitude { get; set; }

    public Bird(string name, int age, string color, int maxAltitude) : base(name, age)
    {
        Color = color;
        MaxAltitude = maxAltitude;
    }

    public override void MakeSound()
    {
        Console.WriteLine($"{Name} чирикает!");
    }

    public override void DisplayInfo()
    {
        base.DisplayInfo();
        Console.WriteLine($"Окрас: {Color}, Максимальная высота полета: {MaxAltitude} м");
    }

    public void Fly()
    {
        Console.WriteLine($"{Name} летит на высоте до {MaxAltitude} метров.");
    }
}

public class Duck : Animal, IFlyable, ISwimmable
{
    public int MaxAltitude { get; set; }
    public bool IsWaterproof { get; set; }

    public Duck(string name, int age, int maxAltitude, bool isWaterproof) : base(name, age)
    {
        MaxAltitude = maxAltitude;
        IsWaterproof = isWaterproof;
    }

    public override void MakeSound()
    {
        Console.WriteLine($"{Name} крякает!");
    }

    public override void DisplayInfo()
    {
        base.DisplayInfo();
        Console.WriteLine($"Максимальная высота полета: {MaxAltitude} м, Водонепроницаемый: {IsWaterproof}");
    }

    public void Fly()
    {
        Console.WriteLine($"{Name} летит на высоте до {MaxAltitude} метров.");
    }

    public void Swim()
    {
        Console.WriteLine($"{Name} плавает в воде. Водонепроницаемость: {IsWaterproof}");
    }
}

public class Cat : Animal
{
    public bool IsIndoor { get; set; }

    public Cat(string name, int age, bool isIndoor) : base(name, age)
    {
        IsIndoor = isIndoor;
    }

    public override void MakeSound()
    {
        Console.WriteLine($"{Name} мяукает!");
    }

    public override void DisplayInfo()
    {
        base.DisplayInfo();
        Console.WriteLine($"Домашняя кошка: {(IsIndoor ? "Да" : "Нет")}");
    }
}
```
