```polyglot-notebook
using System;

// --- ИСПОЛНЯЕМАЯ ЧАСТЬ ---

Console.WriteLine("=== Демонстрация явной реализации интерфейса ===");

// Создание экземпляра Dog
Dog myDog = new Dog("Рекс", 3, "Немецкая овчарка");

// Прямой вызов методов (используется реализация из класса Dog)
myDog.MakeSound();      // Вызовет: "Рекс лает: Гав-гав!"
myDog.Play();           // Вызовет: "Рекс играет."
myDog.Fetch();          // Вызовет: "Рекс приносит мячик."

Console.WriteLine();

// Вызов через интерфейс IAnimal (используется явная реализация)
IAnimal animalInterface = myDog;
animalInterface.Eat();      // Вызовет: "Рекс поел. Уровень энергии: 120"
animalInterface.Sleep();    // Вызовет: "Рекс поспал. Энергия восстановлена."
string name = animalInterface.GetName();
Console.WriteLine($"Имя животного: {name}");

// animalInterface.MakeSound(); // Ошибка компиляции! Метод не определен в интерфейсе
// animalInterface.Fetch();     // Ошибка компиляции! Метод не определен в интерфейсе

Console.WriteLine();
Console.WriteLine("=== Демонстрация с другими животными ===");

// Создание экземпляра Cat
Cat myCat = new Cat("Мурка", 5, true);

// Прямой вызов
myCat.MakeSound();  // "Мурка мяукает: Мяу..."
myCat.Purr();       // "Мурка мурлычет."

// Вызов через интерфейс
IAnimal catInterface = myCat;
catInterface.Eat();
catInterface.Sleep();

Console.WriteLine();

// Создание экземпляра Bird
Bird myBird = new Bird("Кеша", 2, 25.5);

// Прямой вызов
myBird.MakeSound(); // "Кеша чирикает: Чик-чирик!"
myBird.Fly();       // "Кеша летит с размахом крыльев 25.5 см."

// Вызов через интерфейс
IAnimal birdInterface = myBird;
birdInterface.Eat();
birdInterface.Sleep();

Console.WriteLine();
Console.WriteLine("=== Полиморфизм: коллекция животных ===");

// Коллекция объектов типа IAnimal
IAnimal[] animals = new IAnimal[]
{
    new Dog("Бобик", 2, "Дворняга"),
    new Cat("Сима", 4, false),
    new Bird("Попугай", 1, 15.0)
};

foreach (var animal in animals)
{
    animal.Eat();
    animal.Sleep();
    Console.WriteLine(new string('-', 30));
}

// Чтобы вызвать уникальные методы, нужно привести тип
Console.WriteLine("\n=== Вызов уникальных методов после приведения типа ===");
foreach (var animal in animals)
{
    if (animal is Dog dog)
    {
        dog.Fetch();
    }
    else if (animal is Cat cat)
    {
        cat.Purr();
    }
    else if (animal is Bird bird)
    {
        bird.Fly();
    }
}

// --- КЛАССЫ И ИНТЕРФЕЙСЫ ---

// 1. Определение интерфейса IAnimal
public interface IAnimal
{
    void Eat();
    void Sleep();
    string GetName();
}

// 2. Абстрактный базовый класс Animal, реализующий IAnimal
public abstract class Animal : IAnimal
{
    public string Name { get; set; }
    public int Age { get; set; }
    protected double energyLevel;

    public Animal(string name, int age)
    {
        Name = name;
        Age = age;
        energyLevel = 100;
    }

    // Явная реализация методов интерфейса IAnimal
    void IAnimal.Eat()
    {
        energyLevel += 20;
        Console.WriteLine($"{Name} поел. Уровень энергии: {energyLevel}");
    }

    void IAnimal.Sleep()
    {
        energyLevel = 100;
        Console.WriteLine($"{Name} поспал. Энергия восстановлена.");
    }

    string IAnimal.GetName()
    {
        return Name;
    }

    // Абстрактный метод, который должны реализовать все наследники
    public abstract void MakeSound();

    // Неявный метод, доступный напрямую
    public virtual void Play()
    {
        Console.WriteLine($"{Name} играет.");
        energyLevel -= 10;
    }
}

// 3. Производный класс Dog
public class Dog : Animal
{
    public string Breed { get; set; }

    public Dog(string name, int age, string breed) : base(name, age)
    {
        Breed = breed;
    }

    public override void MakeSound()
    {
        Console.WriteLine($"{Name} лает: Гав-гав!");
    }

    public void Fetch()
    {
        Console.WriteLine($"{Name} приносит мячик.");
    }
}

// 4. Производный класс Cat
public class Cat : Animal
{
    public bool IsIndoor { get; set; }

    public Cat(string name, int age, bool isIndoor) : base(name, age)
    {
        IsIndoor = isIndoor;
    }

    public override void MakeSound()
    {
        Console.WriteLine($"{Name} мяукает: Мяу...");
    }

    public void Purr()
    {
        Console.WriteLine($"{Name} мурлычет.");
    }
}

// 5. Производный класс Bird
public class Bird : Animal
{
    public double WingSpan { get; set; }

    public Bird(string name, int age, double wingSpan) : base(name, age)
    {
        WingSpan = wingSpan;
    }

    public override void MakeSound()
    {
        Console.WriteLine($"{Name} чирикает: Чик-чирик!");
    }

    public void Fly()
    {
        Console.WriteLine($"{Name} летит с размахом крыльев {WingSpan} см.");
    }
}
```
