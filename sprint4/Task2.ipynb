```polyglot-notebook
using System;

// --- ИСПОЛНЯЕМАЯ ЧАСТЬ ---

Console.WriteLine("=== Демонстрация внедрения зависимостей ===");

// Создание объектов
IAnimal myDog = new Dog("Рекс", 3, "Немецкая овчарка");
IAnimal myCat = new Cat("Мурка", 5, true);
IAnimal myBird = new Bird("Кеша", 2, 25.5);

// Создание сервисов с внедрением зависимостей
AnimalService dogService = new AnimalService(myDog);
AnimalService catService = new AnimalService(myCat);
AnimalService birdService = new AnimalService(myBird);

// Использование сервисов
Console.WriteLine("--- Работа с собакой ---");
dogService.DisplayInfo();
dogService.CareForAnimal();

Console.WriteLine();
Console.WriteLine("--- Работа с кошкой ---");
catService.DisplayInfo();
catService.CareForAnimal();

Console.WriteLine();
Console.WriteLine("--- Работа с птицей ---");
birdService.DisplayInfo();
birdService.CareForAnimal();

Console.WriteLine();
Console.WriteLine("=== Полиморфизм: коллекция животных ===");

// Коллекция объектов типа IAnimal
IAnimal[] animals = new IAnimal[]
{
    new Dog("Бобик", 2, "Дворняга"),
    new Cat("Сима", 4, false),
    new Bird("Попугай", 1, 15.0),
    new Dog("Шарик", 6, "Лабрадор")
};

foreach (var animal in animals)
{
    AnimalService service = new AnimalService(animal);
    service.DisplayInfo();
    service.CareForAnimal();
    Console.WriteLine(new string('-', 40));
}

Console.WriteLine();
Console.WriteLine("=== Гибкость внедрения зависимостей ===");

// Можно легко заменить реализацию без изменения кода сервиса
IAnimal exoticBird = new Bird("Ара", 10, 80.0);
AnimalService exoticService = new AnimalService(exoticBird);

Console.WriteLine("Работа с экзотической птицей:");
exoticService.DisplayInfo();
exoticService.CareForAnimal();

// --- КЛАССЫ И ИНТЕРФЕЙСЫ ---

// 1. Определение интерфейса IAnimal
public interface IAnimal
{
    void Eat();
    void Sleep();
    string GetName();
}

// 2. Абстрактный базовый класс Animal, реализующий IAnimal
public abstract class Animal : IAnimal
{
    public string Name { get; set; }
    public int Age { get; set; }
    protected double energyLevel;

    public Animal(string name, int age)
    {
        Name = name;
        Age = age;
        energyLevel = 100;
    }

    public virtual void Eat()
    {
        energyLevel += 20;
        Console.WriteLine($"{Name} поел. Уровень энергии: {energyLevel}");
    }

    public virtual void Sleep()
    {
        energyLevel = 100;
        Console.WriteLine($"{Name} поспал. Энергия восстановлена.");
    }

    public string GetName()
    {
        return Name;
    }

    public abstract void MakeSound();

    public virtual void Play()
    {
        Console.WriteLine($"{Name} играет.");
        energyLevel -= 10;
    }
}

// 3. Производный класс Dog
public class Dog : Animal
{
    public string Breed { get; set; }

    public Dog(string name, int age, string breed) : base(name, age)
    {
        Breed = breed;
    }

    public override void Eat()
    {
        Console.WriteLine($"{Name} (собака) жадно ест корм.");
        base.Eat();
    }

    public override void MakeSound()
    {
        Console.WriteLine($"{Name} лает: Гав-гав!");
    }

    public void Fetch()
    {
        Console.WriteLine($"{Name} приносит мячик.");
    }
}

// 4. Производный класс Cat
public class Cat : Animal
{
    public bool IsIndoor { get; set; }

    public Cat(string name, int age, bool isIndoor) : base(name, age)
    {
        IsIndoor = isIndoor;
    }

    public override void Sleep()
    {
        Console.WriteLine($"{Name} (кошка) спит на подоконнике.");
        base.Sleep();
    }

    public override void MakeSound()
    {
        Console.WriteLine($"{Name} мяукает: Мяу...");
    }

    public void Purr()
    {
        Console.WriteLine($"{Name} мурлычет.");
    }
}

// 5. Производный класс Bird
public class Bird : Animal
{
    public double WingSpan { get; set; }

    public Bird(string name, int age, double wingSpan) : base(name, age)
    {
        WingSpan = wingSpan;
    }

    public override void Eat()
    {
        Console.WriteLine($"{Name} клюет зерна.");
        base.Eat();
    }

    public override void MakeSound()
    {
        Console.WriteLine($"{Name} чирикает: Чик-чирик!");
    }

    public void Fly()
    {
        Console.WriteLine($"{Name} летит с размахом крыльев {WingSpan} см.");
    }
}

// 6. Класс сервиса, использующий внедрение зависимостей
public class AnimalService
{
    private readonly IAnimal _animal;

    public AnimalService(IAnimal animal)
    {
        _animal = animal;
    }

    public void CareForAnimal()
    {
        Console.WriteLine($"--- Уход за животным: {_animal.GetName()} ---");

        _animal.Eat();

        if (_animal is Dog dog)
        {
            dog.Fetch();
        }
        else if (_animal is Cat cat)
        {
            cat.Purr();
        }
        else if (_animal is Bird bird)
        {
            bird.Fly();
        }

        _animal.Sleep();
    }

    public void DisplayInfo()
    {
        if (_animal is Animal a)
        {
            Console.WriteLine($"Имя: {a.Name}, Возраст: {a.Age} лет");
        }
    }
}
```
