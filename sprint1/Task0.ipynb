```polyglot-notebook
using System;

// Интерфейс для класса Animal
public interface IAnimal
{
    void MakeSound();
    void Move();
}

// Еще один интерфейс
public interface IFeedable
{
    void Feed(string food);
}

// Базовый класс Animal
public class Animal : IAnimal, IFeedable
{
    // === АТРИБУТЫ (ПОЛЯ) С РАЗНЫМИ МОДИФИКАТОРАМИ ДОСТУПА ===
    
    // Public свойство - доступно из любого места
    public string Name { get; set; }
    
    // Public свойство только для чтения
    public string Species { get; private set; }
    
    // Protected поле - доступно в базовом и производных классах
    protected int age;
    
    // Private поле - доступно только внутри класса Animal
    private string _internalId;
    
    // Private поле для отслеживания состояния
    private bool _isAlive;
    
    // === СТАТИЧЕСКИЕ ЧЛЕНЫ КЛАССА ===
    
    // Статическое свойство для подсчета всех созданных животных
    public static int TotalAnimals { get; private set; }
    
    // Статическое поле (константа)
    public static readonly string Kingdom = "Animalia";
    
    // === КОНСТРУКТОРЫ ===
    
    // Конструктор по умолчанию
    public Animal()
    {
        Name = "Безымянное";
        Species = "Неизвестно";
        age = 0;
        _internalId = Guid.NewGuid().ToString();
        _isAlive = true;
        TotalAnimals++; // Увеличиваем счетчик при создании
        Console.WriteLine($"Создано новое животное (ID: {_internalId})");
    }
    
    // Конструктор с параметрами
    public Animal(string name, string species, int age)
    {
        Name = name;
        Species = species;
        this.age = age;
        _internalId = Guid.NewGuid().ToString();
        _isAlive = true;
        TotalAnimals++;
        Console.WriteLine($"Создано животное: {Name} ({Species})");
    }
    
    // === МЕТОДЫ ===
    
    // Public метод - реализация интерфейса IAnimal
    public virtual void MakeSound()
    {
        Console.WriteLine($"{Name} издает какой-то звук...");
    }
    
    // Public метод - реализация интерфейса IAnimal
    public virtual void Move()
    {
        Console.WriteLine($"{Name} движется...");
    }
    
    // Public метод - реализация интерфейса IFeedable
    public virtual void Feed(string food)
    {
        if (_isAlive)
        {
            Console.WriteLine($"{Name} ест {food}");
        }
        else
        {
            Console.WriteLine($"{Name} не может есть (мертвое)");
        }
    }
    
    // Public метод для получения возраста
    public int GetAge()
    {
        return age;
    }
    
    // Public метод для установки возраста
    public void SetAge(int newAge)
    {
        if (newAge >= 0)
        {
            age = newAge;
            Console.WriteLine($"{Name} теперь {age} лет");
        }
    }
    
    // Protected метод - доступен только в базовом и производных классах
    protected void Sleep()
    {
        Console.WriteLine($"{Name} спит...");
    }
    
    // Private метод - доступен только внутри класса Animal
    private void CheckHealth()
    {
        if (_isAlive)
        {
            Console.WriteLine($"{Name} здорово(a)");
        }
    }
    
    // Статический метод - принадлежит классу, а не экземпляру
    public static void DisplayKingdom()
    {
        Console.WriteLine($"Царство: {Kingdom}");
        Console.WriteLine($"Всего животных создано: {TotalAnimals}");
    }
    
    // Метод для отображения состояния класса
    public void DisplayState()
    {
        Console.WriteLine($"\n=== Состояние животного ===");
        Console.WriteLine($"Имя: {Name}");
        Console.WriteLine($"Вид: {Species}");
        Console.WriteLine($"Возраст: {age}");
        Console.WriteLine($"Живо: {_isAlive}");
        Console.WriteLine($"Внутренний ID: {_internalId}");
        Console.WriteLine($"==========================\n");
    }
}

// === ПРОИЗВОДНЫЙ КЛАСС (для демонстрации protected) ===
public class Dog : Animal
{
    // Private поле только для Dog
    private string _breed;
    
    // Конструктор производного класса
    public Dog(string name, string breed) : base(name, "Собака", 0)
    {
        _breed = breed;
        Console.WriteLine($"Создана собака породы {_breed}");
    }
    
    // Переопределение метода базового класса
    public override void MakeSound()
    {
        Console.WriteLine($"{Name} лает: Гав-гав!");
    }
    
    // Переопределение метода Move
    public override void Move()
    {
        Console.WriteLine($"{Name} бегает на четырех лапах");
    }
    
    // Новый метод, специфичный для Dog
    public void Fetch(string item)
    {
        Console.WriteLine($"{Name} приносит {item}");
        // Вызов protected метода базового класса
        Sleep();
    }
    
    // Метод для отображения породы
    public void DisplayBreed()
    {
        Console.WriteLine($"Порода: {_breed}");
    }
}

// === ДЕМОНСТРАЦИЯ РАБОТЫ ===
Console.WriteLine("=== ДЕМОСТРАЦИЯ КЛАССА ANIMAL ===\n");

// Использование статического метода
Animal.DisplayKingdom();

// Создание объектов с разными конструкторами
Console.WriteLine("\n1. Создание животных:");
Animal animal1 = new Animal(); // Конструктор по умолчанию
Animal animal2 = new Animal("Барсик", "Кошка", 3); // Конструктор с параметрами

// Создание производного класса
Dog dog1 = new Dog("Шарик", "Овчарка");

// Демонстрация public свойств и методов
Console.WriteLine("\n2. Работа с public членами:");
animal2.Name = "Мурзик";
Console.WriteLine($"Имя животного: {animal2.Name}");

// Демонстрация методов
Console.WriteLine("\n3. Вызов методов:");
animal1.MakeSound();
animal2.MakeSound();
dog1.MakeSound(); // Полиморфизм

dog1.Move();
dog1.Fetch("мяч");

// Демонстрация Feed
Console.WriteLine("\n4. Кормление:");
animal2.Feed("рыбу");
dog1.Feed("кость");

// Демонстрация состояния класса
Console.WriteLine("\n5. Отображение состояния:");
animal2.DisplayState();
dog1.DisplayState();

// Демонстрация static членов
Console.WriteLine("\n6. Статические члены:");
Console.WriteLine($"Всего создано животных: {Animal.TotalAnimals}");
Console.WriteLine($"Царство: {Animal.Kingdom}");

// Демонстрация protected (через производный класс)
Console.WriteLine("\n7. Protected члены (доступны в производном классе):");
dog1.DisplayBreed();

Console.WriteLine("\n=== ДЕМОСТРАЦИЯ ЗАВЕРШЕНА ===");
```
