```polyglot-notebook
using System;

// === БАЗОВЫЙ КЛАСС ANIMAL ===
public class Animal
{
    // Статический атрибут
    public static int TotalAnimals { get; private set; }

    // Обычные атрибуты (поля)
    private string _name;
    private int _age;
    private double _weight;

    // Свойства с геттерами и сеттерами
    public string Name
    {
        get { return _name; }
        set 
        { 
            if (!string.IsNullOrEmpty(value))
                _name = value;
        }
    }

    public int Age
    {
        get { return _age; }
        set 
        { 
            if (value >= 0 && value <= 50)
                _age = value;
        }
    }

    public double Weight
    {
        get { return _weight; }
        set 
        { 
            if (value > 0)
                _weight = value;
        }
    }

    // Конструктор
    public Animal(string name, int age, double weight)
    {
        Name = name;
        Age = age;
        Weight = weight;
        TotalAnimals++;
    }

    // Обычные методы класса
    public virtual void MakeSound()
    {
        Console.WriteLine($"{_name} издает звук...");
    }

    public virtual void Eat()
    {
        Console.WriteLine($"{_name} ест...");
    }

    public void Sleep()
    {
        Console.WriteLine($"{_name} спит...");
    }

    // Статический метод
    public static void DisplayTotalAnimals()
    {
        Console.WriteLine($"Всего животных создано: {TotalAnimals}");
    }

    public override string ToString()
    {
        return $"Имя: {_name}, Возраст: {_age}, Вес: {_weight} кг";
    }
}

// === ПРОИЗВОДНЫЙ КЛАСС DOG ===
public class Dog : Animal
{
    private string _breed;
    private bool _isTrained;

    public string Breed
    {
        get { return _breed; }
        set { _breed = value; }
    }

    public bool IsTrained
    {
        get { return _isTrained; }
        set { _isTrained = value; }
    }

    public Dog(string name, int age, double weight, string breed) 
        : base(name, age, weight)
    {
        Breed = breed;
        IsTrained = false;
    }

    public override void MakeSound()
    {
        Console.WriteLine($"{Name} лает: Гав-гав!");
    }

    public override void Eat()
    {
        Console.WriteLine($"{Name} ест собачий корм...");
    }

    public void Fetch()
    {
        if (_isTrained)
            Console.WriteLine($"{Name} приносит мячик!");
        else
            Console.WriteLine($"{Name} не обучен приносить предметы.");
    }

    public void Train()
    {
        _isTrained = true;
        Console.WriteLine($"{Name} теперь обучен!");
    }
}

// === ПРОИЗВОДНЫЙ КЛАСС CAT ===
public class Cat : Animal
{
    private string _color;
    private bool _isIndoor;

    public string Color
    {
        get { return _color; }
        set { _color = value; }
    }

    public bool IsIndoor
    {
        get { return _isIndoor; }
        set { _isIndoor = value; }
    }

    public Cat(string name, int age, double weight, string color, bool isIndoor) 
        : base(name, age, weight)
    {
        Color = color;
        IsIndoor = isIndoor;
    }

    public override void MakeSound()
    {
        Console.WriteLine($"{Name} мяукает: Мяяяу!");
    }

    public override void Eat()
    {
        Console.WriteLine($"{Name} ест кошачий корм...");
    }

    public void Purr()
    {
        Console.WriteLine($"{Name} мурлычет...");
    }

    public void Scratch()
    {
        Console.WriteLine($"{Name} точит когти...");
    }

    public void CheckLifestyle()
    {
        Console.WriteLine($"{Name} - {(_isIndoor ? "домашняя" : "уличная")} кошка");
    }
}

// === ПРОИЗВОДНЫЙ КЛАСС BIRD ===
public class Bird : Animal
{
    private double _wingSpan;
    private bool _canFly;

    public double WingSpan
    {
        get { return _wingSpan; }
        set { _wingSpan = value; }
    }

    public bool CanFly
    {
        get { return _canFly; }
        set { _canFly = value; }
    }

    public Bird(string name, int age, double weight, double wingSpan) 
        : base(name, age, weight)
    {
        WingSpan = wingSpan;
        CanFly = true;
    }

    public override void MakeSound()
    {
        Console.WriteLine($"{Name} поет: Чик-чирик!");
    }

    public override void Eat()
    {
        Console.WriteLine($"{Name} клюет зерна...");
    }

    public void Fly()
    {
        Console.WriteLine($"{Name} {(CanFly ? "летит на высоте..." : "не может летать!")}");
    }

    public void BuildNest()
    {
        Console.WriteLine($"{Name} строит гнездо...");
    }
}

// === ДЕМОСТРАЦИЯ РАБОТЫ КЛАССОВ ===
Console.WriteLine("╔════════════════════════════════════╗");
Console.WriteLine("║  ДЕМОСТРАЦИЯ КЛАССОВ ANIMAL, DOG, CAT, BIRD  ║");
Console.WriteLine("╚════════════════════════════════════╝\n");

// 1. Создание объектов
Console.WriteLine("📋 1. СОЗДАНИЕ ОБЪЕКТОВ:");
Console.WriteLine("------------------------");
Dog dog = new Dog("Шарик", 3, 15.5, "Немецкая овчарка");
Cat cat = new Cat("Мурзик", 2, 4.2, "Рыжий", true);
Bird bird = new Bird("Кеша", 1, 0.5, 30.0);
Console.WriteLine("✓ Созданы: собака, кошка, птица\n");

// 2. Демонстрация свойств (геттеры/сеттеры)
Console.WriteLine("🔧 2. РАБОТА СО СВОЙСТВАМИ:");
Console.WriteLine("---------------------------");
Console.WriteLine($"Собака: {dog.Name}, Порода: {dog.Breed}");
Console.WriteLine($"Кошка: {cat.Name}, Цвет: {cat.Color}, Домашняя: {cat.IsIndoor}");
Console.WriteLine($"Птица: {bird.Name}, Размах крыльев: {bird.WingSpan} см");

// Изменение через сеттер
dog.Age = 4;
Console.WriteLine($"\n✓ Возраст собаки изменен на: {dog.Age}");

// Валидация в сеттере
cat.Age = 100; // Не изменится, т.к. > 50
Console.WriteLine($"✓ Попытка установить возраст 100: реальный возраст = {cat.Age}\n");

// 3. Демонстрация методов базового класса
Console.WriteLine("🎯 3. МЕТОДЫ БАЗОВОГО КЛАССА:");
Console.WriteLine("------------------------------");
Animal genericAnimal = new Animal("Зверь", 5, 10.0);
genericAnimal.MakeSound();
genericAnimal.Eat();
genericAnimal.Sleep();
Console.WriteLine();

// 4. Демонстрация переопределённых методов (полиморфизм)
Console.WriteLine("🔄 4. ПОЛИМОРФИЗМ - ПЕРЕОПРЕДЕЛЕНИЕ МЕТОДОВ:");
Console.WriteLine("-------------------------------------------");
dog.MakeSound();  // Вызывает Dog.MakeSound()
dog.Eat();        // Вызывает Dog.Eat()
Console.WriteLine();

cat.MakeSound();  // Вызывает Cat.MakeSound()
cat.Eat();        // Вызывает Cat.Eat()
Console.WriteLine();

bird.MakeSound(); // Вызывает Bird.MakeSound()
bird.Eat();       // Вызывает Bird.Eat()
Console.WriteLine();

// 5. Специфичные методы производных классов
Console.WriteLine("⭐ 5. СПЕЦИФИЧНЫЕ МЕТОДЫ ПРОИЗВОДНЫХ КЛАССОВ:");
Console.WriteLine("--------------------------------------------");
Console.WriteLine("→ Собака:");
dog.Train();
dog.Fetch();
Console.WriteLine();

Console.WriteLine("→ Кошка:");
cat.Purr();
cat.Scratch();
cat.CheckLifestyle();
Console.WriteLine();

Console.WriteLine("→ Птица:");
bird.Fly();
bird.BuildNest();
Console.WriteLine();

// 6. Демонстрация статического члена
Console.WriteLine("📊 6. СТАТИЧЕСКИЙ ЧЛЕН КЛАССА:");
Console.WriteLine("-------------------------------");
Animal.DisplayTotalAnimals();
Console.WriteLine();

// 7. Демонстрация полиморфизма через массив базового типа
Console.WriteLine("🎭 7. ПОЛИМОРФИЗМ ЧЕРЕЗ МАССИВ БАЗОВОГО ТИПА:");
Console.WriteLine("---------------------------------------------");
Animal[] animals = { dog, cat, bird };
Console.WriteLine("Вызов MakeSound() для всех животных:");
foreach (var animal in animals)
{
    animal.MakeSound(); // Автоматически вызывается правильный метод!
}
Console.WriteLine();

// 8. Демонстрация ToString()
Console.WriteLine("📝 8. ПЕРЕОПРЕДЕЛЕНИЕ ToString():");
Console.WriteLine("----------------------------------");
Console.WriteLine(dog.ToString());
Console.WriteLine(cat.ToString());
Console.WriteLine(bird.ToString());
Console.WriteLine();

// 9. Демонстрация валидации в свойствах
Console.WriteLine("🛡️ 9. ВАЛИДАЦИЯ В СЕТТЕРАХ:");
Console.WriteLine("---------------------------");
Console.WriteLine("Попытка установить отрицательный вес:");
dog.Weight = -10;
Console.WriteLine($"✓ Вес собаки остался: {dog.Weight} кг (отрицательное значение отклонено)");

Console.WriteLine("\nПопытка установить пустое имя:");
cat.Name = "";
Console.WriteLine($"✓ Имя кошки осталось: '{cat.Name}' (пустое значение отклонено)\n");

// 10. Финальный итог
Console.WriteLine("╔════════════════════════════════════╗");
Console.WriteLine("║         ИТОГИ ДЕМОСТРАЦИИ          ║");
Console.WriteLine("╚════════════════════════════════════╝");
Console.WriteLine($"✓ Создано животных всего: {Animal.TotalAnimals}");
Console.WriteLine($"✓ Продемонстрировано:");
Console.WriteLine("  • Структура класса и наследование");
Console.WriteLine("  • Свойства с геттерами/сеттерами");
Console.WriteLine("  • Статические и обычные члены");
Console.WriteLine("  • Переопределение методов (полиморфизм)");
Console.WriteLine("  • Валидация данных в свойствах");
Console.WriteLine("  • Специфичные методы производных классов");
```
