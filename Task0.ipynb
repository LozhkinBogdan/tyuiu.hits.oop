```polyglot-notebook
// Ячейка 1: Инициализация массива и основные операции
int[] numbers = {5, 8, 1, 3, 9, 6, 7, 2};

// 1. Найти минимальное и максимальное значения
int min = numbers.Min();
int max = numbers.Max();
Console.WriteLine($"1. Минимальное: {min}, Максимальное: {max}");

// 2. Сумма всех элементов
int sum = numbers.Sum();
Console.WriteLine($"2. Сумма: {sum}");

// 3. Среднее арифметическое
double average = numbers.Average();
Console.WriteLine($"3. Среднее арифметическое: {average:F2}");

// 4. Сортировка по возрастанию
int[] sorted = numbers.OrderBy(x => x).ToArray();
Console.WriteLine($"4. Отсортированный: {string.Join(", ", sorted)}");

// 5. Проверка наличия числа (статически заданного)
int checkNumber = 6;
bool exists = numbers.Contains(checkNumber);
Console.WriteLine($"5. Число {checkNumber} в массиве: {exists}");
```
