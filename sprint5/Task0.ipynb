```polyglot-notebook
using System;
using System.Collections.Generic;

// --- ИСПОЛНЯЕМАЯ ЧАСТЬ ---

// ===== Работа с List<Animal> =====
Console.WriteLine("=== Работа с List<Animal> ===\n");

List<Animal> animalList = new List<Animal>
{
    new Dog("Рекс", 5, "Овчарка"),
    new Cat("Мурка", 3, true),
    new Bird("Кеша", 2, 25.5),
    new Fish("Немо", 1, "Солёная")
};

Console.WriteLine("Список животных:");
foreach (var animal in animalList)
{
    animal.DisplayInfo();
    animal.MakeSound();
    Console.WriteLine(new string('-', 40));
}

// Вызов специфичных методов через проверку типа
Console.WriteLine("\nСпецифичные действия:");
foreach (var animal in animalList)
{
    if (animal is Dog dog)
        dog.Fetch();
    else if (animal is Cat cat)
        cat.Purr();
    else if (animal is Bird bird)
        bird.Fly();
    else if (animal is Fish fish)
        fish.Swim();
}

// ===== Работа с Dictionary<string, Animal> =====
Console.WriteLine("\n\n=== Работа с Dictionary<string, Animal> ===\n");

Dictionary<string, Animal> animalDictionary = new Dictionary<string, Animal>
{
    { "DOG001", new Dog("Барсик", 4, "Лабрадор") },
    { "CAT002", new Cat("Снежок", 2, false) },
    { "BRD003", new Bird("Попугайчик", 1, 15.0) },
    { "FSH004", new Fish("Золотая рыбка", 3, "Пресная") }
};

Console.WriteLine("Все животные в словаре:");
foreach (var kvp in animalDictionary)
{
    Console.WriteLine($"Ключ: {kvp.Key}");
    kvp.Value.DisplayInfo();
    kvp.Value.MakeSound();
    Console.WriteLine(new string('-', 40));
}

// Безопасный поиск по ключу
Console.WriteLine("\nПоиск по ключу 'CAT002':");
if (animalDictionary.TryGetValue("CAT002", out Animal foundAnimal))
{
    foundAnimal.DisplayInfo();
    foundAnimal.MakeSound();
}
else
{
    Console.WriteLine("Животное не найдено!");
}

Console.WriteLine("\nПоиск по ключу 'XYZ999':");
if (animalDictionary.TryGetValue("XYZ999", out foundAnimal))
{
    foundAnimal.DisplayInfo();
}
else
{
    Console.WriteLine("Животное не найдено!");
}

// Проверка наличия ключа и добавление
Console.WriteLine("\nДобавление нового животного:");
string newKey = "DOG005";
if (!animalDictionary.ContainsKey(newKey))
{
    animalDictionary.Add(newKey, new Dog("Тузик", 6, "Бультерьер"));
    Console.WriteLine($"Животное с ключом {newKey} успешно добавлено!");
}

Console.WriteLine($"\nОбщее количество животных в словаре: {animalDictionary.Count}");

// --- КЛАССЫ ---

// Базовый класс Animal
public class Animal
{
    public string Name { get; set; }
    public int Age { get; set; }

    public Animal(string name, int age)
    {
        Name = name;
        Age = age;
    }

    public virtual void MakeSound()
    {
        Console.WriteLine($"{Name} издаёт звук...");
    }

    public virtual void DisplayInfo()
    {
        Console.WriteLine($"Имя: {Name}, Возраст: {Age} лет");
    }
}

// Производный класс Dog
public class Dog : Animal
{
    public string Breed { get; set; }

    public Dog(string name, int age, string breed) : base(name, age)
    {
        Breed = breed;
    }

    public override void MakeSound()
    {
        Console.WriteLine($"{Name} лает: Гав-гав!");
    }

    public override void DisplayInfo()
    {
        base.DisplayInfo();
        Console.WriteLine($"Порода: {Breed}");
    }

    public void Fetch()
    {
        Console.WriteLine($"{Name} приносит мячик!");
    }
}

// Производный класс Cat
public class Cat : Animal
{
    public bool IsIndoor { get; set; }

    public Cat(string name, int age, bool isIndoor) : base(name, age)
    {
        IsIndoor = isIndoor;
    }

    public override void MakeSound()
    {
        Console.WriteLine($"{Name} мяукает: Мяу-мяу!");
    }

    public override void DisplayInfo()
    {
        base.DisplayInfo();
        Console.WriteLine($"Домашний: {(IsIndoor ? "Да" : "Нет")}");
    }

    public void Purr()
    {
        Console.WriteLine($"{Name} мурлычет...");
    }
}

// Производный класс Bird
public class Bird : Animal
{
    public double Wingspan { get; set; }

    public Bird(string name, int age, double wingspan) : base(name, age)
    {
        Wingspan = wingspan;
    }

    public override void MakeSound()
    {
        Console.WriteLine($"{Name} чирикает: Чик-чирик!");
    }

    public override void DisplayInfo()
    {
        base.DisplayInfo();
        Console.WriteLine($"Размах крыльев: {Wingspan} см");
    }

    public void Fly()
    {
        Console.WriteLine($"{Name} летает!");
    }
}

// Производный класс Fish
public class Fish : Animal
{
    public string WaterType { get; set; }

    public Fish(string name, int age, string waterType) : base(name, age)
    {
        WaterType = waterType;
    }

    public override void MakeSound()
    {
        Console.WriteLine($"{Name} булькает: Буль-буль!");
    }

    public override void DisplayInfo()
    {
        base.DisplayInfo();
        Console.WriteLine($"Тип воды: {WaterType}");
    }

    public void Swim()
    {
        Console.WriteLine($"{Name} плавает!");
    }
}
```
