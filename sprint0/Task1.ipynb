Инкапсуляция

```polyglot-notebook
class Animal
{
    private string name; 
    private int age; 

    public string Name 
    {
        get { return name; }
        set { name = value; }
    }

    public int Age 
    {
        get { return age; }
        set { age = value; }
    }

    public void MakeSound() 
    {
        Console.WriteLine("Животное издаёт звук...");
    }
}

Animal myAnimal = new Animal(); 
myAnimal.Name = "Барсик";
myAnimal.Age = 13;
myAnimal.MakeSound();
Console.WriteLine(myAnimal.Name);
Console.WriteLine(myAnimal.Age);
```

Наследование

```polyglot-notebook
class Animal
{
    private string name; 

    public string Name 
    {
        get { return name; }
        set { name = value; }
    }

    public void Eat() 
    {
        Console.WriteLine($"{name} ест...");
    }
}


class Dog : Animal
{
    public void Bark() 
    {
        Console.WriteLine("Гав-гав!");
    }
}


Animal myAnimal = new Animal(); 
myAnimal.Name = "Кот";
myAnimal.Eat();

Dog myDog = new Dog(); 
myDog.Name = "Шарик"; 
myDog.Eat(); 
myDog.Bark(); 
```

Полиморфизм

```polyglot-notebook
class Animal
{
    public virtual void MakeSound()
    {
        Console.WriteLine("Животное издает звук");
    }
}


class Dog : Animal
{
    public override void MakeSound()
    {
        Console.WriteLine("Собака лает: Гав-гав!");
    }
}


Animal myAnimal = new Animal(); 
Dog myDog = new Dog(); 

myAnimal.MakeSound(); 
myDog.MakeSound();    
```

Абстракция

```polyglot-notebook
abstract class Animal
{
    public abstract void MakeSound(); 

    public void Sleep() 
    {
        Console.WriteLine("Животное спит...");
    }
}


class Dog : Animal
{
    public override void MakeSound() 
    {
        Console.WriteLine("Собака лает: Гав-гав!");
    }
}


Dog myDog = new Dog(); 

myDog.MakeSound(); 
myDog.Sleep();     
```

Композиция

```polyglot-notebook
using System;

public class Heart
{
    public void Beat() => Console.WriteLine("Сердце бьется");
    public void Stop() => Console.WriteLine("Сердце остановилось");
}

public class Animal
{
    private Heart heart;

    public Animal(Heart heart)
    {
        this.heart = heart;
    }

    public void Live()
    {
        heart.Beat();
        
        Console.WriteLine("Животное живет");
        heart.Stop();
    }
}

Heart myHeart = new Heart(); 
Animal myAnimal = new Animal(myHeart); 

myAnimal.Live(); 
```

Агрегация

```polyglot-notebook
using System;


abstract class Animal
{
    public abstract void MakeSound();
}


class Owner
{
    public string Name { get; set; }

    public Owner(string name)
    {
        Name = name;
    }

    public void Feed(Animal animal)
    {
        Console.WriteLine($"{Name} кормит животное...");
        animal.MakeSound();
    }
}


class Cat : Animal
{
    private Owner owner; 

    public Cat(Owner owner)
    {
        this.owner = owner;
    }

    public override void MakeSound()
    {
        Console.WriteLine("Мяу...");
    }
}

// Основная программа
Owner Danil = new Owner("Богдан"); // Создаем владельца
Cat myCat = new Cat(Danil); // Создаем кота с владельцем (агрегация)

myCat.MakeSound(); // Кот мяукает
Danil.Feed(myCat);  // Владелец кормит кота
```
