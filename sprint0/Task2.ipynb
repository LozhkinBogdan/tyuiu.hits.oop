```polyglot-notebook
using System;


public class AnimalBad
{
    private string name;
    
    public void Move()
    {
        Console.WriteLine($"{name} движется...");
    }
    
    public void MakeSound()
    {
        Console.WriteLine($"{name} издает звук...");
    }
    
    public void SaveToDatabase()  // Нарушение SRP!
    {
        Console.WriteLine($"Сохранение {name} в базу данных...");
    }
}


public abstract class Animal
{
    protected string name;
    
    public Animal(string name)
    {
        this.name = name;
    }
    
    public abstract void MakeSound();
    public abstract void Eat();
}


public interface IRepository<T>
{
    void Save(T item);
}


public interface IMovable
{
    void Move();
}


public class Dog : Animal, IMovable
{
    public Dog(string name) : base(name) { }
    
    public override void MakeSound()
    {
        Console.WriteLine($"{name} лает: Гав-гав!");
    }
    
    public override void Eat()
    {
        Console.WriteLine($"{name} ест корм...");
    }
    
    public void Move()
    {
        Console.WriteLine($"{name} бежит...");
    }
}


public class AnimalRepository : IRepository<Animal>
{
    public void Save(Animal animal)
    {
        Console.WriteLine($"Сохранение животного {animal} в базу данных...");
    }
}


Dog myDog = new Dog("Шарик");
myDog.MakeSound();  
myDog.Eat();        
myDog.Move();       

AnimalRepository repository = new AnimalRepository();
repository.Save(myDog);  
```

Open/Closed Principle

```polyglot-notebook
using System;

public interface IMakeSound
{
    void MakeSound();
}


public class Animal : IMakeSound
{
    protected string name;
    
    public Animal(string name)
    {
        this.name = name;
    }
    
    public virtual void MakeSound()
    {
        Console.WriteLine($"{name} издает звук...");
    }
}


public class Dog : Animal
{
    public Dog(string name) : base(name) { }
    
    public override void MakeSound()
    {
        Console.WriteLine($"{name} лает: Гав-гав!");
    }
    
    public void Fetch()
    {
        Console.WriteLine($"{name} приносит мяч...");
    }
}

public class Cat : Animal
{
    public Cat(string name) : base(name) { }
    
    public override void MakeSound()
    {
        Console.WriteLine($"{name} мяукает: Мяу-мяу!");
    }
    
    public void Scratch()
    {
        Console.WriteLine($"{name} точит когти...");
    }
}


public class Bird : Animal
{
    public Bird(string name) : base(name) { }
    
    public override void MakeSound()
    {
        Console.WriteLine($"{name} чирикает: Чик-чирик!");
    }
    
    public void Fly()
    {
        Console.WriteLine($"{name} летит...");
    }
}


public class AnimalShelter
{
    public void MakeAllSounds(IMakeSound[] animals)
    {
        foreach (var animal in animals)
        {
            animal.MakeSound(); // Работает с любым животным!
        }
    }
}


Dog dog = new Dog("Шарик");
Cat cat = new Cat("Барсик");
Bird bird = new Bird("Кеша");


IMakeSound[] animals = { dog, cat, bird };

AnimalShelter shelter = new AnimalShelter();
shelter.MakeAllSounds(animals);

// Дополнительные методы конкретных классов
dog.Fetch();
cat.Scratch();
bird.Fly();
```

Liskov Substitution Principle

```polyglot-notebook
using System;


public class AnimalBad
{
    public virtual void Move()
    {
        Console.WriteLine("Животное движется...");
    }
    
    public virtual void MakeSound()
    {
        Console.WriteLine("Животное издает звук...");
    }
}

public class FishBad : AnimalBad
{
    public override void Move()
    {
        Console.WriteLine("Рыба плывет..."); // Разное поведение!
    }
    
    public void Swim()
    {
        Console.WriteLine("Рыба плавает в воде...");
    }
}

public class BirdBad : AnimalBad
{
    public override void Move()
    {
        Console.WriteLine("Птица летит..."); 
    }
    
    public void Fly()
    {
        Console.WriteLine("Птица летает в небе...");
    }
}


public void DisplayMovementInfo(AnimalBad animal)
{
    animal.Move(); 
}


public interface ILandAnimal
{
    void Walk();
    void MakeSound();
}

public interface IFlyingAnimal
{
    void Fly();
    void MakeSound();
}

public interface ISwimmingAnimal
{
    void Swim();
}


public class Dog : ILandAnimal
{
    private string name;
    
    public Dog(string name)
    {
        this.name = name;
    }
    
    public void Walk()
    {
        Console.WriteLine($"{name} идет по земле...");
    }
    
    public void MakeSound()
    {
        Console.WriteLine($"{name} лает: Гав-гав!");
    }
}

public class Eagle : IFlyingAnimal
{
    private string name;
    
    public Eagle(string name)
    {
        this.name = name;
    }
    
    public void Fly()
    {
        Console.WriteLine($"{name} летает в небе...");
    }
    
    public void MakeSound()
    {
        Console.WriteLine($"{name} кричит: Кррр-кррр!");
    }
}

public class Shark : ISwimmingAnimal
{
    private string name;
    
    public Shark(string name)
    {
        this.name = name;
    }
    
    public void Swim()
    {
        Console.WriteLine($"{name} плавает в океане...");
    }
}



public void DisplayLandAnimalMovement(ILandAnimal animal)
{
    animal.Walk(); // Всегда ожидаем ходьбу!
    animal.MakeSound();
}

public void DisplayFlyingAnimalMovement(IFlyingAnimal animal)
{
    animal.Fly(); 
    animal.MakeSound();
}


Dog dog = new Dog("Шарик");
Eagle eagle = new Eagle("Орел");
Shark shark = new Shark("Акула");


DisplayLandAnimalMovement(dog);      
DisplayFlyingAnimalMovement(eagle);  


dog.Walk();
eagle.Fly();
shark.Swim();
```

Interface Segregation Principle

```polyglot-notebook
using System;

// === ИСПРАВЛЕНИЕ ISP (правильно) ===
// Разделяем большой интерфейс на маленькие конкретные интерфейсы

public interface IAnimal
{
    void Move();
}

public interface ICanMakeSound
{
    void MakeSound();
}

public interface ICanFly
{
    void Fly();
}

public interface ICanSwim
{
    void Swim();
}

// === Классы реализуют только нужные интерфейсы ===

// Собака: движется и издает звуки
public class Dog : IAnimal, ICanMakeSound
{
    private string name;
    
    public Dog(string name)
    {
        this.name = name;
    }
    
    public void Move()
    {
        Console.WriteLine($"{name} бежит...");
    }
    
    public void MakeSound()
    {
        Console.WriteLine($"{name} лает: Гав-гав!");
    }
}

// Птица: движется, издает звуки и летает
public class Bird : IAnimal, ICanMakeSound, ICanFly
{
    private string name;
    
    public Bird(string name)
    {
        this.name = name;
    }
    
    public void Move()
    {
        Console.WriteLine($"{name} передвигается...");
    }
    
    public void MakeSound()
    {
        Console.WriteLine($"{name} поет: Чик-чирик!");
    }
    
    public void Fly()
    {
        Console.WriteLine($"{name} летит...");
    }
}

// Рыба: движется и плавает (не летает!)
public class FishCorrect : IAnimal, ICanSwim
{
    private string name;
    
    public FishCorrect(string name)
    {
        this.name = name;
    }
    
    public void Move()
    {
        Console.WriteLine($"{name} движется в воде...");
    }
    
    public void Swim()
    {
        Console.WriteLine($"{name} плывет...");
    }
}

// Акула: движется, плавает (не издает звуки как другие!)
public class Shark : IAnimal, ICanSwim
{
    private string name;
    
    public Shark(string name)
    {
        this.name = name;
    }
    
    public void Move()
    {
        Console.WriteLine($"{name} движется...");
    }
    
    public void Swim()
    {
        Console.WriteLine($"{name} быстро плывет...");
    }
}

// === Использование ===
Console.WriteLine("=== ДЕМОСТРАЦИЯ ПРИНЦИПА ISP ===\n");

Dog dog = new Dog("Шарик");
Bird bird = new Bird("Кеша");
FishCorrect fish = new FishCorrect("Немо");
Shark shark = new Shark("Акула");

// Каждое животное использует только свои методы
Console.WriteLine("1. Собака:");
dog.Move();
dog.MakeSound();

Console.WriteLine("\n2. Птица:");
bird.Move();
bird.MakeSound();
bird.Fly();

Console.WriteLine("\n3. Рыба:");
fish.Move();
fish.Swim();

Console.WriteLine("\n4. Акула:");
shark.Move();
shark.Swim();

// Можно работать с животными через конкретные интерфейсы
Console.WriteLine("\n5. Работа через интерфейс IAnimal:");
IAnimal[] animals = { dog, bird, fish, shark };
foreach (var animal in animals)
{
    animal.Move(); // Все животные могут двигаться
}

Console.WriteLine("\n6. Работа через интерфейс ICanFly:");
ICanFly flyingAnimal = bird;
flyingAnimal.Fly();

Console.WriteLine("\n7. Работа через интерфейс ICanSwim:");
ICanSwim[] swimmingAnimals = { fish, shark };
foreach (var animal in swimmingAnimals)
{
    animal.Swim();
}

Console.WriteLine("\n=== ПРИНЦИП ISP ПРОДЕМОНСТРИРОВАН ===");
```

Dependency Inversion Principle

```polyglot-notebook
using System;


public class Meat
{
    public void Consume()
    {
        Console.WriteLine("Съедено мясо...");
    }
}

public class DogBad
{
    private Meat _food;

    public DogBad()
    {
        _food = new Meat(); 
    }

    public void Eat()
    {
        _food.Consume();
    }
}



public interface IFood
{
    void Consume();
}


public class MeatDip : IFood
{
    public void Consume()
    {
        Console.WriteLine("Съедено мясо...");
    }
}

public class Grass : IFood
{
    public void Consume()
    {
        Console.WriteLine("Съедена трава...");
    }
}


public class Animal
{
    private IFood _food;


    public Animal(IFood food)
    {
        _food = food;
    }

    public void Eat()
    {
        _food.Consume();
    }
}


IFood meat = new MeatDip();
IFood grass = new Grass();


Animal lion = new Animal(meat); 
Animal rabbit = new Animal(grass); 

lion.Eat();    
rabbit.Eat();  
```
