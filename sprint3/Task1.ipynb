```polyglot-notebook
using System;
using System.Collections.Generic;

// --- ИСПОЛНЯЕМАЯ ЧАСТЬ ---

Console.WriteLine("=== ДЕМОНСТРАЦИЯ ПЕРЕГРУЗКИ МЕТОДОВ (OVERLOADING) ===");
Animal myAnimal = new Animal("Неизвестное животное", 1, "серый");

// Вызываем разные версии метода Move() в зависимости от типа параметра
myAnimal.Move(50);                    // Вызывает Move(int)
myAnimal.Move("на север");            // Вызывает Move(string)
myAnimal.Move(10.5, 2.0);             // Вызывает Move(double, double)
Console.WriteLine();

Console.WriteLine("=== ДЕМОНСТРАЦИЯ ПЕРЕКРЫТИЯ МЕТОДОВ (OVERRIDING) И ПОЛИМОРФИЗМА ===");
// Создаем список базового типа, но заполняем его производными объектами
List<Animal> animals = new List<Animal>
{
    new Dog("Шарик", 3, "черный", "Овчарка"),
    new Cat("Мурка", 5, "рыжая", true),
    new Bird("Кеша", 2, "зеленый", 0.8),
    new Animal("Неизвестное животное", 1, "серый")
};

foreach (var animal in animals)
{
    // При вызове MakeSound() будет выполнена версия, соответствующая
    // реальному типу объекта (полиморфизм).
    animal.MakeSound();
    
    // Вызываем обычный метод базового класса
    animal.Sleep();
    
    Console.WriteLine(); // Пустая строка для разделения
}

Console.WriteLine("=== ДЕМОНСТРАЦИЯ СПЕЦИФИЧНЫХ МЕТОДОВ И ЧАСТИЧНОГО ПЕРЕКРЫТИЯ ===");
// Чтобы вызвать специфичные методы или проверить частичное перекрытие,
// нужно привести тип к конкретному классу
foreach (var animal in animals)
{
    if (animal is Dog dog)
    {
        dog.Fetch();
        // Проверим, как работает частично перекрытый метод Move(int)
        dog.Move(40); // Здесь вызовется overridden метод, который сначала вызовет base.Move(40)
    }
    else if (animal is Cat cat)
    {
        cat.Purr();
        cat.Move(30); // Здесь вызовется полностью замененный метод
    }
    else if (animal is Bird bird)
    {
        bird.Fly();
        bird.Move("в южном направлении"); // Здесь вызовется overridden метод Move(string)
    }
}

// --- КЛАССЫ ---

// === БАЗОВЫЙ КЛАСС ANIMAL ===
public class Animal
{
    public string Name { get; set; }
    public int Age { get; set; }
    public string Color { get; set; }

    public Animal(string name, int age, string color)
    {
        Name = name;
        Age = age;
        Color = color;
    }

    // --- ПЕРЕГРУЗКА МЕТОДОВ (Overloading) ---
    // Методы помечены как virtual, чтобы их можно было переопределять в наследниках

    // Версия 1: принимает скорость
    public virtual void Move(int speed)
    {
        Console.WriteLine($"[Animal] {Name} движется со скоростью {speed} км/ч.");
    }

    // Версия 2: принимает направление
    public virtual void Move(string direction)
    {
        Console.WriteLine($"[Animal] {Name} движется в направлении: {direction}.");
    }

    // Версия 3: принимает расстояние и время (не переопределяется, virtual не обязателен)
    public void Move(double distance, double timeHours)
    {
        double avgSpeed = distance / timeHours;
        Console.WriteLine($"[Animal] {Name} прошел {distance} км за {timeHours} ч. Средняя скорость: {avgSpeed:F1} км/ч.");
    }

    // --- ВИРТУАЛЬНЫЙ МЕТОД ДЛЯ ПЕРЕКРЫТИЯ (Overriding) ---
    public virtual void MakeSound()
    {
        Console.WriteLine($"[Animal] {Name} издает какой-то звук.");
    }

    public void Sleep()
    {
        Console.WriteLine($"[Animal] {Name} спит.");
    }
}

// === ПРОИЗВОДНЫЙ КЛАСС DOG ===
public class Dog : Animal
{
    public string Breed { get; set; }

    public Dog(string name, int age, string color, string breed) : base(name, age, color)
    {
        Breed = breed;
    }

    public override void MakeSound()
    {
        Console.WriteLine($"[Dog] {Name} ({Breed}) лает: Гав-гав!");
    }

    // Теперь можно переопределять, так как в базовом классе метод virtual
    public override void Move(int speed)
    {
        base.Move(speed);
        Console.WriteLine($"[Dog] {Name} радостно виляет хвостом при беге!");
    }

    public void Fetch()
    {
        Console.WriteLine($"[Dog] {Name} приносит палку.");
    }
}

// === ПРОИЗВОДНЫЙ КЛАСС CAT ===
public class Cat : Animal
{
    public bool IsIndoor { get; set; }

    public Cat(string name, int age, string color, bool isIndoor) : base(name, age, color)
    {
        IsIndoor = isIndoor;
    }

    public override void MakeSound()
    {
        string location = IsIndoor ? "дома" : "на улице";
        Console.WriteLine($"[Cat] {Name} мяукает: Мяу-мяу! (живет {location})");
    }

    // Теперь можно переопределять
    public override void Move(int speed)
    {
        Console.WriteLine($"[Cat] {Name} крадется со скоростью {speed} км/ч... очень тихо.");
    }

    public void Purr()
    {
        Console.WriteLine($"[Cat] {Name} мурлычет.");
    }
}

// === ПРОИЗВОДНЫЙ КЛАСС BIRD ===
public class Bird : Animal
{
    public double Wingspan { get; set; }

    public Bird(string name, int age, string color, double wingspan) : base(name, age, color)
    {
        Wingspan = wingspan;
    }

    public override void MakeSound()
    {
        Console.WriteLine($"[Bird] {Name} щебечет: Чик-чирик! (размах крыльев: {Wingspan} м)");
    }

    // Теперь можно переопределять Move(string)
    public override void Move(string direction)
    {
        Console.WriteLine($"[Bird] {Name} летит в направлении: {direction}, взмахивая крыльями.");
    }

    public void Fly()
    {
        Console.WriteLine($"[Bird] {Name} парит высоко в небе.");
    }
}
```
